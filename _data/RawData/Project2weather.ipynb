```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
api_key = ""

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities_weather_in_US.csv"

# Range of latitudes and longitudes
lat_range = (19.50139, 64.85694)
lng_range = (-161.75583 , -68.01197)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=19.50139, high=64.85694, size=1500)
print(lats)
lngs = np.random.uniform(low=-161.75583, high=-68.01197, size=1500)
print(lngs)
lat_lngs = zip(lats, lngs)
```

```python
# Identify nearest city for each lat, lng combination
cntr = 0
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lats[cntr], lngs[cntr]).city_name
    cntr = cntr + 1
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
cities
```

```python
# Starting URL for Weather Map API Call

collected_data = []

base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
#Hide the api key
def hide_key(base_url):
    return base_url[0:47] + base_url[100:]
#Loop thru the cities and check if the data exists before appending
for city in cities:
    run_url = base_url + "&q=" + city
#    print(run_url)
    print(hide_key(run_url))
    r = requests.get(run_url)
    if r.status_code == 200:        
        response = requests.get(run_url).json()
#    print(response)
        collected_data.append(response)
```

```python
len(collected_data)
```

```python
# Use list comprehension to gather necessary data series
import datetime
lat_data = [data.get("coord").get("lat") for data in collected_data]
lng_data = [data.get("coord").get("lon") for data in collected_data]
temp_data = [data.get("main").get("temp") for data in collected_data]
min_temp = [data.get("main").get("temp_min") for data in collected_data]
max_temp = [data.get("main").get("temp_max") for data in collected_data]
sea_level = [data.get("main").get("sea_level") for data in collected_data]
grnd_level = [data.get("main").get("grnd_level") for data in collected_data]
humid_data = [data.get("main").get("humidity") for data in collected_data]
cloud_data =[data.get("clouds").get("all") for data in collected_data]
wind_data = [data.get("wind").get("speed") for data in collected_data]
wind_degrees = [data.get("wind").get("deg") for data in collected_data]
validcities = [data.get("name") for data in collected_data]
countryList = [data.get("sys").get("country") for data in collected_data]
dateList = str(datetime.datetime.today()).split()[0]
#dateList = "2018-11-12"
```

```python
# Create data frame with aggregated data
weather_data = pd.DataFrame({"cityName":validcities,
                            "country":countryList,
                            "lat":lat_data,
                            "lng":lng_data,
                            "temp":temp_data,
                             "minTemp": min_temp,
                             "maxTemp": max_temp,
                             "seaLevel":sea_level,
                             "grndLevel":grnd_level,
                            "humidity":humid_data,
                            "cloudiness":cloud_data,
                            "winds":wind_data,
                            "windDegrees":wind_degrees,
                            "date": dateList})

# re order columns 
weather_data = weather_data[["cityName","country","lat","lng","temp","minTemp","maxTemp","seaLevel","grndLevel", "humidity","cloudiness","winds","windDegrees","date"]]
weather_data_new = weather_data[weather_data.country == "US"]
weather_data_new.reset_index(inplace=True, drop=True)
```

```python
# Save weather data to csv file
output_data_file = "cities_weather_in_US.csv"
#output_data_file = "test_1.csv"
weather_data_new.to_csv(output_data_file, mode='a', header=False, encoding="utf-8", index=False)
```
